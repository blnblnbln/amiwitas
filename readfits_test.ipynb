```python
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from astropy.wcs import WCS
```

```python
fits_image = fits.open('ngc1313_49.5621_-66.5330.fits')
image = fits_image[0].data
```

```python
RA = fits_image[0].header['CRVAL1']
DEC = fits_image[0].header['CRVAL2']
```

```python
wcs_dict = {
    'CTYPE1'  : 'RA---TAN',           #                               
    'CTYPE2'  : 'DEC--TAN',           # TANgent plane                                  
    'CRVAL1'  :              49.5621, # Reference RA                                   
    'CRVAL2'  :              -66.533, # Reference Dec                                  
    'CRPIX1'  :                128.5, # Reference x                                    
    'CRPIX2'  :                128.5, # Reference y                                    
    'CD1_1'   : -0.00222222222222222, # CD matrix                                      
    'CD1_2'   :                   0., # CD matrix                                      
    'CD2_1'   :                   0., # CD matrix                                      
    'CD2_2 '  :  0.00222222222222222, # CD matrix                                      
    'IMAGEW'  :                  256, # Image width                                    
    'IMAGEH'  :                  256 # Image height 
    }  
```

```python
wcs_dict = WCS(wcs_dict)
```

```python
g = image[0]
r = image[1]
i = image[2]
z = image[3]
```

```python
fig, ax = plt.subplots(1, 4, figsize=(15, 5), subplot_kw={'projection': wcs_dict}, sharex=True, sharey=True)

im0 = ax[0].imshow(g, cmap='Greens', norm=LogNorm(1e-2, 10))
im1 = ax[1].imshow(r, cmap='Reds', norm=LogNorm(1e-2, 10))
im2 = ax[2].imshow(i, cmap='Purples', norm=LogNorm(1e-2, 10))
im3 = ax[3].imshow(z, cmap='PuBu', norm=LogNorm(1e-2, 10))

cbar_0 = fig.colorbar(im0, ax=ax[0], orientation='vertical', shrink=0.6)
cbar_1 = fig.colorbar(im1, ax=ax[1], orientation='vertical', shrink=0.6)
cbar_2 = fig.colorbar(im2, ax=ax[2], orientation='vertical', shrink=0.6)
cbar_3 = fig.colorbar(im3, ax=ax[3], orientation='vertical', shrink=0.6)


ax[0].set_xlabel('RA')
ax[0].set_ylabel('Dec')
ax[1].set_xlabel('RA')
ax[1].set_ylabel('Dec')
ax[2].set_xlabel('RA')
ax[2].set_ylabel('Dec')
ax[3].set_xlabel('RA')
ax[3].set_ylabel('Dec')
```
